### Concatenation

```python
df1 = df_sub[['AGE', 'TAX', 'INDUS', 'B']].sample(n=4)
df2 = df_sub[['AGE', 'TAX', 'INDUS', 'B']].sample(n=4)
df3 = df_sub[['AGE', 'TAX', 'INDUS', 'B']].sample(n=4)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
cat1 = pd.concat([df1, df2, df3], axis=0)
cat1
```

```python
cat2 = pd.concat([df1, df2, df3], axis=1)
cat2
```

```python
print(cat2.isnull().sum())
print('We finally have a few missing values for our previous exercise')
```

### Merging using common keys

```python
pd.merge(df1, df2, on='TAX', how='inner')
```

```python
pd.merge(df1, df2, on='TAX', how='inner').drop_duplicates()
```

```python
pd.merge(df2, df3, on='TAX', how='inner')
```

```python
pd.merge(df2, df3, on='TAX', how='inner').drop_duplicates()
```
