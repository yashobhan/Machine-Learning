# Random Forest on the LendingClub Data

Working with LendingClub data - LendingClub connects people who need money with people who have money (borrowers & investors).

Creating a model that predics the risk of lending money to someone with a wide range of credit-related data.

We will predict whether or not the borrower paid back the loan in full.

Data description:
- credit.policy: 1 if the customer meets the credit criteria, 0 otherwise
- purpose: purpose of the loan
- int.rate: interest rate on the loan
- installment: montly installments owed by the borrower
- log.natural.inc: natural log of the self-reported annual income of the borrower
- dti: debt-to-income ratio of the borrower
- fico: FICO score of the borrower
- dats.with.cr.line: No. of days the borrower has the credit line
- revol.bal: the borrower's revolving balance (amt unpaid at the end of credit line)
- revol.util: the borrower's revolving line utilization rate
- inq.last.6.mths: borrower's no. of inquiries by the creditor in last 6 months
- delinq.2yrs: the number of times the borrower had been 30+ days past the due date on a payment in past 2 years
- pub.rec: the borrower's no. of public derogatory public records
- not.fully.paid: the quantity of interest for classification - whether borrower paid back or not

## Data Setup

### Categorical features

The purpose column is categorical. We transform them using dummy variables (one-hot encode them) so sklearn can process them

```python
df_final = pd.get_dummies(df, ['purpose'], drop_first=True)
df_final.head()
```

### Train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_final.drop('not.fully.paid', axis=1)
y = df_final['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
X_train.shape, X_test.shape
```

## Training Models

### Training a Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier(criterion='gini')
dtree
```

```python
dtree.fit(X_train, y_train)
```

### Making Predictions

```python
preds = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test, preds))
```

```python
cm = confusion_matrix(y_test, preds)
print(cm)
```

```python
print('Accuracy of prediction: ', round((cm[0, 0] + cm[1, 1]) / cm.sum(), 3))
```

### Training a Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_jobs=-1, n_estimators=500)
```

```python
rf.fit(X_train, y_train)
```

### Making predictions

```python
rf_preds = rf.predict(X_test)
```

```python
cr = classification_report(y_test, preds)
print(cr)
```

```python
cm = confusion_matrix(y_test, preds)
print(cm)
```

### Looping increasing number of trees in Random Forest & checking accuracy

```python
nsimu = 30
accuracy = [0] * nsimu
n_tree = [0] * nsimu

for i in range(1, nsimu):
    rf = RandomForestClassifier(n_estimators=i*5, 
                                 min_samples_split=10,
                                 max_depth=None, 
                                 criterion='gini')
    
    rf.fit(X_train, y_train)
    preds = rf.predict(X_test)
    cm = confusion_matrix(y_test, preds)
    accuracy[i] = (cm[0, 0] + cm[1, 1]) / cm.sum()
    n_tree[i] = i * 5
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x=n_tree[1:nsimu], y=accuracy[1:nsimu], s=60, c='blue')
plt.title('No. of trees in RF vs Prediction acc')
plt.xlabel('No of trees')
plt.ylabel('Prediction acc')
plt.grid()
```
