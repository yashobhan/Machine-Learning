## GAN

```python
class GAN:
    def __init__(self, img_rows, img_cols, channels):
        self.img_rows = img_rows
        self.img_cols = img_cols
        self.channels = channels
        self.img_shape = (self.img_rows, self.img_cols, self.channels)
        
        optimizer = tf.keras.optimizers.Adam(lr=0.0002, beta_1=0.5)
        
        
        ## Discriminator
        self.discriminator = self.build_discriminator()
        self.discriminator.compile(loss='binary_crossentropy',
                                   optimizer=optimizer,
                                   metrics=['accuracy'])
        
        
        ## Generator
        self.generator = self.build_generator()
        self.generator.compile(loss='binary_crossentropy', 
                               optimizer=optimizer)
        
        ## Generator takes random noise as input
        z = tf.keras.layers.Input(shape=(self.img_rows, self.img_cols, self.channels))
        img = self.generator(z)
        
        ## We only have to train the generator
        self.discriminator.trainable = False
        
        ## valid takes generated images as inputs and determines validity (discriminates)
        valid = self.discriminator(img)
        
        ## The combined model (GAN) takes
        ## noise as input => generates images => determines validity
        
        self.gan = tf.keras.models.Model(z, valid)
        self.gan.compile(loss='binary_crossentropy',
                         optimizer=optimizer)
        
        
    def build_generator(self):
        noise_shape = (self.img_rows, self.img_cols, self.channels)
        
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Conv2D(128, (3, 3), input_shape=noise_shape, padding='same'))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        model.add(tf.keras.layers.BatchNormalization(momentum=0.8))
        
        model.add(tf.keras.layers.UpSampling2D())
        
        model.add(tf.keras.layers.Conv2D(64, (3, 3), padding='same'))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        model.add(tf.keras.layers.BatchNormalization(momentum=0.8))
        
        model.add(tf.keras.layers.Flatten())
        
        model.add(tf.keras.layers.Dense(128))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        model.add(tf.keras.layers.BatchNormalization(momentum=0.8))
        
        model.add(tf.keras.layers.Dense(256))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        model.add(tf.keras.layers.BatchNormalization(momentum=0.8))
        
        model.add(tf.keras.layers.Dense(512))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        model.add(tf.keras.layers.BatchNormalization(momentum=0.8))
        
        model.add(tf.keras.layers.Dense(np.prod(self.img_shape), activation='tanh'))
        
        model.add(tf.keras.layers.Reshape(self.img_shape))
        
        model.summary()
        
        noise = tf.keras.layers.Input(shape=noise_shape)
        img = model(noise)
        
        return tf.keras.models.Model(noise, img)
    
    
    
    def build_discriminator(self):
        img_shape = (self.img_rows, self.img_cols, self.channels)
        
        model = tf.keras.models.Sequential()
        model.add(tf.keras.layers.Conv2D(128, (3, 3), input_shape=img_shape, padding='same'))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        
        model.add(tf.keras.layers.Conv2D(64, (3, 3), padding='same'))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        
        model.add(tf.keras.layers.Flatten(input_shape=img_shape))
        
        model.add(tf.keras.layers.Dense(256))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        
        model.add(tf.keras.layers.Dense(128))
        model.add(tf.keras.layers.LeakyReLU(alpha=0.2))
        
        model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
        
        model.summary()
        
        img = tf.keras.layers.Input(shape=img_shape)
        validity = model(img)
        
        return tf.keras.models.Model(img, validity)
    
    
    
    def get_image(self, img_path, width, height, mode):
        image = Image.open(img_path)
        
        if image.size != (width, height):
            face_width = face_height = 108
            j = (image.size[0] - face_width) // 2
            i = (image.size[1] - face_width) // 2
            
            image = image.crop([j, i, j + face_width, i + face_height])
            image = image.resize((width, height))
            
        return np.array(image.convert(mode))
    
    
    def get_batch(self, image_files, width, height, mode):
        data_batch = np.array([self.get_image(sample_file, width, height, mode) for sample_file in image_files])
        
        return data_batch
    
    
    def train(self, epochs, data_batch=5000, batch_size=32, save_interval=100):
        DATA_DIR = '../../Resources/data/gan/celeba-dataset/img_align_celeba/img_align_celeba/'
        
        X_train = self.get_batch(glob(os.path.join(DATA_DIR, '*.jpg'))[:data_batch], self.img_rows, self.img_cols, 'RGB')
        
        ## Rescaling to -1, 1
        X_train = (X_train.astype(np.float32) - 127.5) / 127.5
        
        half_batch = int(batch_size / 2)
        
        ## Logging
        d_loss_logs_r = []
        d_loss_logs_f = []
        g_loss_logs = []
        
        for epoch in range(epochs):
            '''
                TRAIN DISCRIMINATOR
            '''
            
            idx = np.random.randint(0, X_train.shape[0], half_batch)
            imgs = X_train[idx]
            
            noise = np.random.normal(0, 1, (half_batch, self.img_rows, self.img_cols, self.channels))
            
            ## Generate half-batch of new images
            gen_imgs = self.generator.predict(noise)
            
            ## Train the discriminator
            d_loss_real = self.discriminator.train_on_batch(imgs, np.ones((half_batch, 1)))
            d_loss_fake = self.discriminator.train_on_batch(gen_imgs, np.zeros((half_batch, 1)))
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            
            
            
            '''
                TRAIN GENERATOR
            '''
            
            noise = np.random.normal(0, 1, (batch_size, self.img_rows, self.img_cols, self.channels))
            
            ## The generator wants the discriminator to label the generated samples as valid (ones)
            valid_y = np.array([1] * batch_size)
            
            ## Train the generator
            g_loss = self.gan.train_on_batch(noise, valid_y)
            
            
            if epoch % (save_interval // 2) == 0:
                print('* Epoch: {}, Discriminator Loss: {}, Discriminator Acc: {}, Generator Loss: {}'.format(epoch, d_loss[0], 100*d_loss[1], g_loss))
            
            d_loss_logs_r.append([epoch, d_loss[0]])
            d_loss_logs_f.append([epoch, d_loss[1]])
            g_loss_logs.append([epoch, g_loss])
            
            d_loss_logs_r_a = np.array(d_loss_logs_r)
            d_loss_logs_f_a = np.array(d_loss_logs_f)
            g_loss_logs_a = np.array(g_loss_logs)
            
            if epoch % save_interval == 0:
                self.save_imgs(epoch)
                     
                plt.plot(d_loss_logs_r_a[:,0], d_loss_logs_r_a[:,1], label="Discriminator Loss - Real")
                plt.plot(d_loss_logs_f_a[:,0], d_loss_logs_f_a[:,1], label="Discriminator Loss - Fake")
                plt.plot(g_loss_logs_a[:,0], g_loss_logs_a[:,1], label="Generator Loss")
                plt.xlabel('Epochs')
                plt.ylabel('Loss')
                plt.legend()
                plt.title('Variation of losses over epochs')
                plt.grid(True)
                plt.show()
            
            
    def save_imgs(self, epoch):
        r, c = 3, 3
        noise = np.random.normal(0, 1, (r * c, self.img_rows, self.img_cols, self.channels))
        gen_imgs = self.generator.predict(noise)

        # Rescale images 0 - 1
        gen_imgs = (1/2.5) * gen_imgs + 0.5

        fig, axs = plt.subplots(r, c)
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i,j].imshow(gen_imgs[cnt, :,:,:])
                axs[i,j].axis('off')
                cnt += 1
        fig.savefig("generations/vanilla_gan/%d.png" % epoch)
        plt.close()
```

```python
img_rows, img_cols = 48, 48
channels = 3

gan = GAN(img_rows, img_cols, channels)
```

```python
gan.train(epochs=2000, batch_size=16, data_batch=5000)
```
